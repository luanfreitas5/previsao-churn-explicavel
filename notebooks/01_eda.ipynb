# 01 — Análise Exploratória (EDA)

Exploração das tabelas Olist e da distribuição do alvo de churn (não recompra).
Lógica de produção vive em `src/`; este notebook é apenas exploração.

```python
from warnings import filterwarnings

import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from src.config.paths import get_paths
from src.config.settings import load_settings
from src.data.loader import load_olist_tables
from src.features.rfm import build_churn_features

filterwarnings("ignore")
RANDOM_SEED = 42
rng = np.random.default_rng(seed=RANDOM_SEED)
rng.normal()
plt.rcParams["figure.dpi"] = 120
sns.set_theme(style="whitegrid")
```

```python
settings = load_settings()
paths = get_paths()
tables = load_olist_tables(paths.data_raw)
tables.orders.head()
```

```python
features = build_churn_features(
    orders=tables.orders,
    customers=tables.customers,
    order_items=tables.order_items,
    payments=tables.payments,
    reviews=tables.reviews,
    churn=settings.churn,
)
print(f"Clientes: {features.height} | Taxa de churn: {features['churn'].mean():.1%}")
features.head()
```

```python
pdf = features.to_pandas()
fig, ax = plt.subplots(figsize=(8, 5))
sns.histplot(data=pdf, x="recency_days", hue="churn", bins=40, ax=ax)
ax.set_title("Recência por classe de churn")
ax.set_xlabel("Dias desde a última compra")
plt.tight_layout()
```
